```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,180)
y=np.cos(x*np.pi/180)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
n=1024
x=np.random.normal(0,1,n)
y=np.random.normal(0,1,n)
T=np.arctan2(y,x)
plt.scatter(x,y,s=75,c=T,alpha=0.5)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```
